```python
import pandas as pd

import pandas as pd
import os

def table_from_folder(folder_path):
    dataframes = {}
    for file in os.listdir(folder_path):
        if file.endswith('.csv'):
            # Read CSV file into a dataframe
            df = pd.read_csv(os.path.join(folder_path, file), index_col=0)
            
            # Extracting part of the filename after the underscore
            name_part = file.split('_')[-1].split('.')[0]
            
            # Saving the dataframe with the extracted name part as the key
            dataframes[name_part] = df
    import matplotlib.pyplot as plt

    # Data
    values = [dataframes[key].loc["accuracy", "precision"] for key in dataframes.keys()]
    labels = [key for key in dataframes.keys()]

    bar_width = 0.8  # Adjust the width of the bars

    # Determining the highest value
    max_value = max(values)
    colors = ['red' if v == max_value else 'blue' for v in values]

    # Creating the bar plot with custom colors and width
    bars = plt.bar(labels, values, color=colors, width=bar_width)

    # Adding title and labels
    plt.title('Accuracy of different ML-Methods')
    plt.xlabel('Method')
    plt.ylabel('Accuracy')

    plt.xticks(rotation=10)  # Rotates labels to
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.3f}', ha='center', va='bottom')


    # Display the plot
    plt.show()


    
    return dataframes



# Now you can access the dataframes using the extracted names as keys
# For example, to access a dataframe with key 'xyz', use dataframes['xyz']
```

```python
# Specify the folder path here
folder_path = 'test_vgg_without_fc'
dataframes = table_from_folder(folder_path)
```
